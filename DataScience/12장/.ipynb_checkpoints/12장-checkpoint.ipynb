```python
#12.5
import csv

f = open('/Users/park_minwoo/Downloads/weather.csv')   #경로
data = csv.reader(f)
header = next(data)
max_wind = 0.0

for row in data:
    if row[3] == '':
        wind = 0
    else:
        wind = float(row[3])
    if max_wind < wind:
        max_wind = wind
        
print(f"지난 10년간 울릉도의 최대 풍속은 {max_wind} m/s")
```

```python
#Lab 12-1
import csv
import matplotlib.pyplot as plt

f = open('/Users/p경기도ark_minwoo/Downloads/weather.csv')   #경로
data = csv.reader(f)
header = next(data)

monthly_wind = [0 for x in range(12)]
days_counted = [0 for x in range(12)]

for row in data:
    month = int(row[0][5:7])
    if row[3] != '':
        wind = float(row[3])
        monthly_wind[month-1] += wind
        days_counted[month-1] += 1

for i in range(12):
    monthly_wind[i] /= days_counted[i]
    
plt.plot(monthly_wind, "hotpink")
plt.show()
f.close()
```

```python
#Lab 12-2
import datetime as dt

weather = pd.read_csv('/Users/park_minwoo/Downloads/weather.csv')
monthly = [None for x in range(12)]
monthly_wind = [0 for x in range(12)]
weather['month'] = pd.DatetimeIndex(weather['일시']).month

for i in range(12):
    monthly[i] = weather[weather['month'] == i+1]
    monthly_wind[i] = monthly[i].mean()['평균풍속']
    
plt.plot(monthly_wind, "red")
plt.show()
```
