```python
n1 = int(input(""))
n2 = int(input(""))
li1 =[]
li2 =[]

for i in range(2,n1+1):
    if n1%i == 0:
        li1.append(i)
        
for i in range(2,n2+1):
    if n2%i == 0:
        li2.append(i)
        
li1 = set(li1)
li2 = set(li2)

print(f"{n1}과 {n2}의 최대공약수는 {max(li1.intersection(li2))}")
```

```python
#LAB 9-1

import string

alp = string.ascii_uppercase
caiser = alp[3:]+alp[:3]

def cipher(a):
    idx = alp.index(a)
    return caiser[idx]

sec = input("문장을 입력하시오 : ")
print("암호화된 문장 : ",end="")

for ch in sec:
    if ch in alp:
        print(cipher(ch), end="")
    else:
        print(ch, end="")
        
print()
```

```python
#9-3

import random, string

src_str = string.ascii_letters + '0123456789'

n_digits = int(input("몇 자리의 비밀번호를 원하십니까? : "))

otp = ''

for i in range(n_digits):
    idx = random.randrange(len(src_str))
    otp += src_str[idx]
print(otp)
```
