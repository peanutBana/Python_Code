```python
import matplotlib.pyplot as plt

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3]

plt.title("GDP per capita")
plt.plot(years, gdp, color="blue", marker="o", linestyle="solid")    #(x_list, y_list, prop)

plt.ylabel("dollars")
#plt.savefig("gdp_per_capita.png",dpi=600)  png이미지로 저장
plt.show()
```

```python
#11.1
import matplotlib.pyplot as plt
import numpy as np

plt.title("numbers")
rand_y = np.random.randn(1000)
rand_x = np.arange(1000)
plt.plot(rand_x,rand_y,marker = 'o',color="red")


plt.axis([0,1000,-3,3])

plt.show()
```

```python
x = [x for x in range(-20,20)]
y1 = [t*2 for t in x]
y2 = [t**2 + 5 for t in x]
y3 = [-t**2 -5 for t in x]

plt.plot(x,y1,'r--', x,y2,'g^-', x,y3,'b*:')    #좌표/색상/라인타입
plt.axis([-30,30,-30,30])
plt.show()
```

```python
x = [x for x in range(20)]
y = [x**2 for x in range(20)]
z = [x**3 for x in range(20)]

plt.plot(x,x, label="linear")    #라벨 기제
plt.plot(x,y, label="quadratic")
plt.plot(x,z, label="qubic")
plt.legend()    #default 위치에 범례를 표시

plt.xlabel('x label')
plt.ylabel('y label')
plt.title("My Plot")
plt.show()
```

```python
#11.2
import numpy as np
import matplotlib.pyplot as plt



x = [x for x in np.arange(0,2,0.01)]
y = [x**2 for x in np.arange(0,2,0.01)]
z = [x**3 for x in np.arange(0,2,0.01)]
k = [2**x for x in np.arange(0,2,0.01)]

plt.plot(x,x, label="linear")
plt.plot(x,y, label="quadratic")
plt.plot(x,z, label="qubic")
plt.plot(x,k, label ="power")
#plt.axis([0,2,0,8])
plt.legend()


plt.xlabel('x label')
plt.ylabel('y label')
plt.title("My Plot")
plt.show()
```

```python
#LAB 11-2
import math
import matplotlib.pyplot as plt

x=[]
y=[]

for angle in range(360):
    x.append(angle)
    y.append(math.sin(math.radians(angle)))
    
plt.plot(x,y)
plt.title("SIN WAVE")
plt.show()print(x)
```

```python
#11.3
import math
x=[]
y=[]

for angle in range(360):
    x.append(angle)
    y.append(math.cos(math.radians(angle)))
    
plt.plot(x,y,"r-")
plt.title("COS WAVE")
plt.show()
```

```python
##막대 그래프
import matplotlib.pyplot as plt

years = [1950,1960,1970,1980,1990,2000,2010]
gdp = [76.0,80.0,257.0,1686.0,6505,11865.3,22105.3]

plt.bar(range(len(years)),gdp)

plt.title("GDP per capita")
plt.ylabel("dollars")

plt.xticks(range(len(years)),years)
plt.show()
```

```python
#다중 막대그래프

years = [1965,1975,1985,1995,2005,2015]
ko = [130,650,2450,11600,17790,27250]
jp = [890,5120,11500,42130,40560,38780]
ch = [100,200,290,540,1760,7940]

x_range = range(len(years))
plt.bar(x_range,jp,width = 0.25)
plt.bar(x_range,ko,width = 0.25)
plt.bar(x_range,ch,width = 0.25)
plt.xticks(x_range,years)
plt.show()
```

```python
years = [1965,1975,1985,1995,2005,2015]
x_range = np.arange(len(years))
plt.bar(x_range+0.0, ko, width = 0.25)
plt.bar(x_range+0.3, jp, width = 0.25)
plt.bar(x_range+0.6, ch, width = 0.25)
```

```python
#lab 11-3
mul1, sigma1 = 10,2
mul2, sigma2 = -6,3

s_Gauss = np.random.randn(10000)
Gauss1 = mul1 + sigma1 * s_Gauss
Gauss2 = mul2 + sigma2 * s_Gauss

plt.figure(figsize=(10,6))
plt.hist(s_Gauss, bins = 50, alpha=0.4)
plt.hist(Gauss1, bins = 50, alpha=0.4)
plt.hist(Gauss2, bins = 50, alpha=0.4)
plt.show()
```
