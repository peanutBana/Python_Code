```python
import numpy as np

x = np.linspace(0,12*np.pi,200)
cos_y = np.cos(x)*20
sin_y = np.sin(x)*x

plt.plot(x,cos_y,linestyle="--",color="blue")
plt.plot(x,sin_y,linestyle="-",color="red")
plt.show()
```

```python
import random

arr=[]

for i in range(30):
    x = random.randint(0,50)
    y = random.randint(0,50)
    arr.append([x,y])

arr_x=[]
arr_y=[]

for i in arr:
    arr_x.append(i[0])
    arr_y.append(i[1])

plt.scatter(arr_x,arr_y)
plt.show()
```

```python
import matplotlib.pyplot as plt

d = [[77,25],[78,28],[85,29],[83,30],[73,21],[77,22],[73,17],[80,35]]
s = [[75,56],[77,57],[86,50],[86,53],[79,60],[83,53],[83,49],[88,61]]
m = [[34,22],[38,25],[38,19],[41,30],[30,21],[37,24],[41,28],[35,18]]

fig, ax = plt.subplots(1,3,figsize=(20,4))
dx = []
dy = []
sx = []
sy = []
mx = []
my = []
for i in d:
    dx.append(i[0])
    dy.append(i[1])
for i in s:
    sx.append(i[0])
    sy.append(i[1])
for i in m:
    mx.append(i[0])
    my.append(i[1])

ax[0].scatter(dx,dy,color="red",marker="o")
ax[1].scatter(sx,sy,color="blue",marker="^")
ax[2].scatter(mx,my,color="green",marker="s")

plt.setp(ax[0:],xlabel="Length")   #전체 요소 라벨 설정
plt.setp(ax[0:],ylabel="Height")

ax[0].set_title("Dacshound size")  #타이틀 각각 설정
ax[1].set_title("Samoyed size")
ax[2].set_title("Maltese size")
plt.show()
```
